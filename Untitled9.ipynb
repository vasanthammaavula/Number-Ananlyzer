# Block-B quality tables : Missingness and **cardinality**

```python
print("\nMISSING (%):\n", df.isna().mean().sort_values(ascending=False).round(3).head(10))
```

```python
df = df.copy()
# drop rows that are useless for numeric EDA
df = df.dropna(subset=['price', 'quantity'])

# remake total_price only where both exist, but don't overwrite existing
missing_tp_mask = df['total_price'].isna() & df['price'].notna() & df['quantity'].notna()
df.loc[missing_tp_mask, 'total_price'] = df['price'] * df['quantity']

print("Rows remaining:", len(df))
df.head()
```

```python
df.info()
```

```python
obj_cols = [c for c in df.columns if df[c].dtype == "object"]
print("\nCARDINALITY (object):\n", pd.Series({c: df[c].nunique() for c in obj_cols}).sort_values(ascending=False))
```

# Block C â€” Distribution Plots (numeric & categorical)*italicized text*


**Numeric: price** **bold text**

```python
plt.figure();
df["price"].dropna().hist(bins=30)
plt.title("Price Distribution");
plt.xlabel("price");
plt.ylabel("count")
```

```python
plt.figure();
df["price"].dropna().plot(kind="box")
plt.title("Price Boxplot")
```

**Categorical**: top cities

```python
top_city = df["city"].value_counts().head(10)
print("\nTOP CITIES:\n", top_city)
```

```python
plt.figure();
top_city.plot(kind="bar")
plt.title("Top Cities by Count");
plt.xlabel("city");
plt.ylabel("count")
```

# Block D â€” Relationships (quick pivot teaser)**bold text**

```python
if {"city","product"}.issubset(df.columns):
    metric = "price" if "price" in df.columns else None
    if metric:
        pivot_ct = df.pivot_table(index="city", columns="product", values=metric, aggfunc="mean")
        print("\nAVG PRICE: city Ã— product (top 5 rows)\n", pivot_ct.head())
```

# Block E â€” Correlation **Matrix**

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if len(num_cols) >= 2:
    corr = df[num_cols].corr(method="pearson")
    print("\nCORR (Pearson):\n", corr.round(3))
```

```python
if len(num_cols) >= 2:
    plt.figure(); plt.imshow(corr, interpolation="nearest")
    plt.title("Correlation Heatmap (Pearson)")
    plt.xticks(range(len(num_cols)), num_cols, rotation=90); plt.yticks(range(len(num_cols)), num_cols)
    plt.colorbar(); plt.tight_layout()
```

# Block F â€” Segment Pivot (city Ã— revenue)**bold text**

```python
if {"city","price","quantity"}.issubset(df.columns):
    df["total_price"] = (df["price"] * df["quantity"]).round(2)
    pivot_rev = df.pivot_table(index="city", values="total_price", aggfunc="sum").sort_values("total_price", ascending=False)
    print("\nREVENUE BY CITY:\n", pivot_rev.head(10))
```

# Block G â€” Target Analysis (class split)**bold text**

```python
if "returned" in df.columns:
    print("\nTARGET SPLIT (returned):\n", df["returned"].value_counts(normalize=True).round(3))
```

# Block H â€” Insight **Export**

```python
insights = []
if "price" in df.columns and df["price"].notna().any():
    insights.append(f"Price median ~ {df['price'].median():.0f}; right tail visible â†’ consider log-transform/winsorizing.")
if {"city","price","quantity"}.issubset(df.columns):
    top_city = df["city"].value_counts().idxmax()
    insights.append(f"Top city by volume: {top_city} â€” prioritize inventory/promo checks.")
if "returned" in df.columns:
    insights.append(f"Return baseline: {df['returned'].mean():.1%} â€” measure uplift by city/product.")

with open("eda_insights.txt", "w", encoding="utf-8") as f:
    f.write("EDA Insights (quick summary)\n- " + "\n- ".join(insights))
print("\nSaved: eda_insights.txt  | Charts: hist/box/cities/corr")
```
